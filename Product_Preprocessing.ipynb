# Product Column Preprocessing

## name 컬럼

- 일단 정확하게 '뉴발란스 992 그레이'라고 적어놓은 제품만 추출함.
- 인덱스는 항상 초기화 해주어야한다.

```python
product = bunkae_df[bunkae_df['name'].str.contains("뉴발란스 992 그레이")] # 제품명을 정확히 포함하는 결과 추출
product = product.reset_index(drop=True)
```

```python
product['name']
```

**제품명에서 파생하여 brand, product_name, color 가져옴.**

```python
product['brand'] = '뉴발란스' # 브랜드 입력.
```

```python
product['product_name'] = '뉴발란스 992 그레이' # 제품명 입력.
```

```python
product['color'] = '그레이' # 색상 입력.
```

#### size 처리
- name컬럼에서 숫자만 뽑아낸뒤, 992숫자를 지운다.
- 이상한값들이 많아서 이는 나중에 없애줘야한다.
- 숫자로 변환해서 150~350 사이숫자만 냅두고, 나머지는 Null값으로 둔다.

```python
import re
# name에서 숫자만 추출
number = []
for i in range(0, len(product)):
    number.append(re.sub(r'[^0-9]','',product['name'].values[i]))
product["size"] = pd.DataFrame(number)
```

```python
# 뉴발란스만 특수상황 : 992라는 숫자가 name에서 포함되서 그럼. 제품마다 상황이 다를 것임.
# name에서 992 제거
size = []
for i in range(0, len(product)):
    size.append(re.sub('992','',product['size'][i]))
product["size"] = pd.DataFrame(size)
```

```python
# 1단계
def size (x):
    return x[0:3]
```

```python
product['size'] = product['size'].apply(size)
```

```python
# # 2단계
# int 변환 후 150 ~ 350 사이값
product['size'] = product['size'].replace('','0') #string
product['size'] = product['size'].astype(int)
product['size'][(product['size'] >= 200) & (product['size'] <= 350)].value_counts().sort_index()
```

## price 컬럼
- 정발가 : 259,000원
- 뉴발란스 992 그레이 크림 거래 범위 기준 : 200,000 이상 500,000 이하
- 이상한 값들 제거해야함

```python
product = product[(product['price'] >= 200000) & (product['price'] <= 500000)] # 시중가격 범위 입력
```

```python
product['release_price'] = 259000 # 정발가입력
```

## location 컬럼
- 시/도/구 등 컬럼을 나눠주려고 한다.

```python
product['location'].head()
```

```python
set(product['location']) # nan존재
```

```python
product['location'] = product['location'].fillna('NULL')
```

```python
def first_split(x):
    first = x.find(' ')
    return x[:first]

def second_split(x):
    first = x.find(' ')
    second = x.find(' ', first+1)
    return x[first+1:second]

def third_split(x):
    first = x.find(' ')
    second = x.find(' ', first+1)
    third = x.find(' ', second+1)
    return x[second+1:]
```

```python
product['location1'] = product['location'].apply(first_split)
product['location2'] = product['location'].apply(second_split)
product['location3'] = product['location'].apply(third_split)
```
