```python
## 번개장터 코드
import pandas as pd
import numpy as np
import json
import time
import requests

bunjang_df = pd.DataFrame()

info = {
    'referer' : 'https://m.bunjang.co.kr/',
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36' 
    }

for i in range(0, 10):
    
    seed = np.random.randint(100)
    np.random.seed(seed)
    a = np.random.randint(5)
    time.sleep(a)
    
    url = f'https://api.bunjang.co.kr/api/1/find_v2.json?q=%EB%82%98%EC%9D%B4%ED%82%A4+%EB%8D%A9%ED%81%AC%EB%A1%9C%EC%9A%B0+%EB%B2%94%EA%B3%A0%EB%9E%98&order=date&page={i}&request_id=20211110132819&stat_device=w&n=100&stat_category_required=1&req_ref=search&version=4'
    
    print(f'{i}페이지 크롤링중...')
    
    resp = requests.get(url, headers=info)
    
    if resp.status_code == requests.codes.ok:
        data = json.loads(resp.text)
        df = pd.DataFrame(data['list'])
        bunjang_df = pd.concat([df, bunjang_df])
    
    else:
        print(f'{page}페이지 요청이 잘못되었습니다.')
        pass

print("크롤링 완료.")
```

```python
'''
name column 에서 숫자만 추출하여 
대략적인 size data 구함
'''

## string data 에서 숫자(사이즈) 추출하는 코드 ##

import re

size = []

# df index 범위만큼 for문 돌림 #
for i in range(0, len(bunjang_df)):
    size.append(re.sub(r'[^0-9]','',bunjang_df['name'].values[i]))

size_df = pd.DataFrame(size)
size_df.columns = ['size']
size
```
