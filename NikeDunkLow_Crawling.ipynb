```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import json
import time
```

```python
# keyword = '나이키 덩크 로우'
# 크롤링 시기 : 2021-11-10 (약)18:30

df = pd.DataFrame()

info = {
    'referer' : 'https://m.bunjang.co.kr/',
    'user-agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/95.0.4638.69 Safari/537.36'
    }

for i in range(0, 110):
    
    seed = np.random.randint(100)
    np.random.seed(seed)
    a = np.random.randint(3)
    time.sleep(a)
    
    url = f'https://api.bunjang.co.kr/api/1/find_v2.json?q=%EB%82%98%EC%9D%B4%ED%82%A4+%EB%8D%A9%ED%81%AC+%EB%A1%9C%EC%9A%B0&order=date&page={i}&request_id=20211110183323&stat_uid=77854434&token=c7533d5959054b9c882ac4276988d849&stat_device=w&n=100&stat_category_required=1&req_ref=search&version=4'
    print(f'{i}회차 크롤링 중...')
    resp = requests.get(url, headers=info)
    
    if resp.status_code == requests.codes.ok:
        data = json.loads(resp.text)
        df_prior = pd.DataFrame(data['list'])
        df = pd.concat([df, df_prior])
        
    else:
        print(f'{i}번째 요청이 잘못 되었습니다.')
        pass

print('크롤링 완료!!')

df
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
# "나이키 덩크 로우" Raw Data 저장
df.to_csv('./NIKEDunkLow_raw.csv')
```
